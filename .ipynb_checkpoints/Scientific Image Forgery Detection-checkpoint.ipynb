#### Section 2 - Features Extraction

```python
# FORENSIC FEATURE EXTRACTION FUNCTIONS
# Each function computes one aspect of the image's forensic fingerprint

def ela_stats(img, quality=90):
    """
    Error Level Analysis (ELA)
    
    HOW IT WORKS:
      1. Re-save image at fixed JPEG quality (90%)
      2. Measure pixel-level differences between original and re-saved
      3. Tampered regions show higher differences (inconsistent JPEG history)
    
    WHY IT DETECTS FORGERIES:
      - Authentic images: uniform JPEG compression throughout
      - Forged images: spliced regions have different compression history
      - Copy-moved patches: show elevated ELA at boundaries
    
    Parameters
    ----------
    img : numpy.ndarray
        BGR image from cv2.imread()
    quality : int
        JPEG quality for recompression (90 = good balance)
    
    Returns
    -------
    tuple of (float, float, float)
        (ELA mean, ELA std, ELA max)
        Higher values = more manipulation evidence
    """
    try:
        # Encode as JPEG at specified quality
        _, encoded = cv2.imencode('.jpg', img, [cv2.IMWRITE_JPEG_QUALITY, quality])
        
        # Decode back to image
        recompressed = cv2.imdecode(encoded, cv2.IMREAD_COLOR)
        
        # Calculate absolute pixel difference
        # Original - Recompressed = compression artifact level
        ela = np.abs(img.astype(np.float32) - recompressed.astype(np.float32))
        
        # Statistics of the difference map
        return float(ela.mean()), float(ela.std()), float(ela.max())
    except:
        return 0.0, 0.0, 0.0
```

```python
def noise_residual_stats(img):
    """
    SRM-Style High-Pass Noise Residual
    
    HOW IT WORKS:
      1. Convert image to greyscale
      2. Apply Gaussian blur (removes low-frequency content)
      3. Subtract blur from original (keeps only high-frequency noise)
      4. Measure noise statistics
    
    WHY IT DETECTS FORGERIES:
      - Authentic images: uniform noise pattern throughout
      - Spliced regions: noise doesn't match surrounding area
      - Copy-move boundaries: sharp noise discontinuities
      - Retouched areas: reduced noise (over-smoothed)
    
    Parameters
    ----------
    img : numpy.ndarray
        BGR image from cv2.imread()
    
    Returns
    -------
    tuple of (float, float)
        (Noise mean, Noise std)
        Inconsistent values indicate tampering
    """
    try:
        # Convert to greyscale for noise analysis
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)
        
        # Apply Gaussian blur (kernel size 5×5, sigma=1.2)
        # This removes high-frequency details, leaving smooth base
        blurred = cv2.GaussianBlur(gray, (5, 5), 1.2)
        
        # Residual = Original - Blurred = noise + texture
        residual = np.abs(gray - blurred)
        
        return float(residual.mean()), float(residual.std())
    except:
        return 0.0, 0.0
```

```python
def shannon_entropy(img):
    """
    Shannon Entropy (Information Complexity)
    
    HOW IT WORKS:
      1. Convert to greyscale
      2. Calculate histogram (pixel value distribution)
      3. Compute information entropy: H = -Σ p(i) * log2(p(i))
    
    WHY IT DETECTS FORGERIES:
      - Forged images have higher entropy (more complex pixel patterns)
      - Manipulation adds new information not present in original
      - Copy-move increases redundancy but local entropy varies
    
    INTERPRETATION:
      - Low entropy (0-4): Smooth, uniform images
      - Medium entropy (4-6): Natural textures
      - High entropy (6-8): Very complex, noisy images
    
    Parameters
    ----------
    img : numpy.ndarray
        BGR image from cv2.imread()
    
    Returns
    -------
    float
        Shannon entropy in bits (0-8 for 8-bit images)
        Higher = more information complexity
    """
    try:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Calculate histogram: 256 bins for 8-bit image
        hist, _ = np.histogram(gray, bins=256, range=(0, 256))
        
        # Convert to probability distribution
        # Add tiny epsilon (1e-10) to avoid log(0)
        prob = hist / (hist.sum() + 1e-10)
        
        # Shannon entropy formula
        # H = -Σ p(i) * log2(p(i))
        entropy = -np.sum(prob * np.log2(prob + 1e-10))
        
        return round(float(entropy), 4)
    except:
        return 0.0
```

```python
def fft_stats(img):
    """
    FFT Frequency Domain Analysis
    
    HOW IT WORKS:
      1. Convert to greyscale
      2. Apply 2D Fast Fourier Transform
      3. Measure energy distribution in frequency domain
    
    WHY IT DETECTS FORGERIES:
      - Copy-move creates periodic repetition
      - Periodic patterns → peaks in frequency spectrum
      - Center energy = low-frequency content (smooth regions)
      - Edge energy = high-frequency content (textures, edges)
    
    WHAT WE MEASURE:
      - Center energy: Energy in DC component + low frequencies
      - Total energy: Average magnitude across all frequencies
    
    Parameters
    ----------
    img : numpy.ndarray
        BGR image from cv2.imread()
    
    Returns
    -------
    tuple of (float, float)
        (Center energy, Total energy) - both log-scaled
        Anomalous patterns indicate potential copy-move
    """
    try:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        H, W = gray.shape
        
        # 2D FFT (transforms spatial domain → frequency domain)
        fft = np.fft.fft2(gray)
        
        # Shift zero-frequency component to center
        fft_shifted = np.fft.fftshift(fft)
        
        # Magnitude spectrum (log scale for visualization)
        magnitude = np.log1p(np.abs(fft_shifted))
        
        # Center energy: 20×20 region around DC component
        center_region = magnitude[H//2-10:H//2+10, W//2-10:W//2+10]
        center_energy = center_region.mean()
        
        # Total energy: average across entire spectrum
        total_energy = magnitude.mean()
        
        return round(float(center_energy), 4), round(float(total_energy), 4)
    except:
        return 0.0, 0.0
```

```python
def extract_all_features(filepath, label):
    """
    Master function: Extract ALL features from one image
    
    This is called once per image (5000 times total)
    Returns a dictionary with all extracted features
    
    Parameters
    ----------
    filepath : str
        Full path to image file
    label : str
        'authentic' or 'forged'
    
    Returns
    -------
    dict or None
        Dictionary with 20+ features, or None if image fails to load
    """
    # Try to load the image
    img = cv2.imread(filepath)
    
    if img is None:
        # Image failed to load (corrupted file)
        return None
    
    # --- SPATIAL DIMENSIONS -----------------------------------------------
    H, W, C = img.shape  # Height, Width, Channels
    pixels = H * W
    megapixels = pixels / 1_000_000
    aspect_ratio = W / H
    
    # --- COLOUR CHANNELS --------------------------------------------------
    # OpenCV loads as BGR, convert to RGB for standard analysis
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Split into individual channels
    R = img_rgb[:, :, 0].astype(np.float32)
    G = img_rgb[:, :, 1].astype(np.float32)
    B = img_rgb[:, :, 2].astype(np.float32)
    
    # Per-channel statistics
    r_mean, r_std = float(R.mean()), float(R.std())
    g_mean, g_std = float(G.mean()), float(G.std())
    b_mean, b_std = float(B.mean()), float(B.std())
    
    # --- GREYSCALE ANALYSIS -----------------------------------------------
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)
    
    # Brightness = mean pixel value (0-255)
    brightness = float(gray.mean())
    
    # Contrast = standard deviation of pixel values
    # Higher std = more variation in brightness = higher contrast
    contrast = float(gray.std())
    
    # --- FORENSIC SIGNALS --------------------------------------------------
    entropy = shannon_entropy(img)
    noise_m, noise_s = noise_residual_stats(img)
    ela_m, ela_s, ela_x = ela_stats(img)
    fft_c, fft_t = fft_stats(img)
    
    # --- RETURN ALL FEATURES AS DICTIONARY ---------------------------------
    return {
        # Identity
        "filename": os.path.basename(filepath),
        "filepath": filepath,
        "label": label,
        
        # Dimensions
        "width": W,
        "height": H,
        "channels": C,
        "pixels": pixels,
        "megapixels": round(megapixels, 4),
        "aspect_ratio": round(aspect_ratio, 4),
        
        # Greyscale statistics
        "brightness": round(brightness, 2),
        "contrast": round(contrast, 2),
        "entropy": entropy,
        
        # RGB channel means
        "r_mean": round(r_mean, 2),
        "r_std": round(r_std, 2),
        "g_mean": round(g_mean, 2),
        "g_std": round(g_std, 2),
        "b_mean": round(b_mean, 2),
        "b_std": round(b_std, 2),
        
        # Forensic signals
        "noise_mean": round(noise_m, 4),
        "noise_std": round(noise_s, 4),
        "ela_mean": round(ela_m, 4),
        "ela_std": round(ela_s, 4),
        "ela_max": round(ela_x, 2),
        "fft_center": fft_c,
        "fft_total": fft_t,
    }
```

```python
# MAIN PROCESSING LOOP
# Process all 5000 images and extract features

print("="*70)
print("  SECTION 2 — FEATURE EXTRACTION")
print("="*70)
print(f"\n  Processing {len(image_list)} images...")
print(f"  This will take approximately {len(image_list)*0.12:.0f} seconds (20~25 min)\n")

# Store results
results = []
errors = 0
start_time = time.time()

# Progress tracking
if HAS_TQDM:
    # Use fancy progress bar if tqdm is installed
    iterator = tqdm(image_list, desc="Extracting features", unit="img")
else:
    # Manual progress updates
    iterator = image_list

# Process each image
for i, (filepath, label) in enumerate(iterator):
    
    # Manual progress for non-tqdm users
    if not HAS_TQDM and i % 100 == 0:
        elapsed = time.time() - start_time
        pct = (i + 1) / len(image_list) * 100
        eta = (elapsed / (i + 1)) * (len(image_list) - i - 1) if i > 0 else 0
        print(f"  [{i+1:>5}/{len(image_list)}]  {pct:>5.1f}%  "
              f"Elapsed: {elapsed:>5.0f}s  ETA: {eta:>5.0f}s", end='\r')
    
    # Extract features
    features = extract_all_features(filepath, label)
    
    if features:
        results.append(features)
    else:
        errors += 1

# Calculate timing
elapsed = time.time() - start_time
avg_time = elapsed / max(len(results), 1)

print(f"\n\n{'='*70}")
print(f"  EXTRACTION COMPLETE")
print(f"{'='*70}")
print(f"\n Processed: {len(results)} images")
print(f"  Failed:    {errors} images")
print(f"  Time:      {elapsed:.1f}s ({avg_time:.2f}s per image)")
print(f"{'='*70}\n")
```

```python
# =========================================================================
# HELPER FUNCTIONS — Consistent Styling
# =========================================================================

def style_ax(ax, title=None, xlabel=None, ylabel=None):
    """
    Apply consistent dark theme styling to a matplotlib axis.
    
    This function sets:
      - Background colour to PANEL
      - Tick colours to MUTED
      - Spine (border) colours to GRID
      - Gridlines in GRID colour
      - Title, xlabel, ylabel if provided
    
    Parameters
    ----------
    ax : matplotlib.axes.Axes
        The axis to style
    title : str, optional
        Chart title
    xlabel : str, optional
        X-axis label
    ylabel : str, optional
        Y-axis label
    """
    ax.set_facecolor(PANEL)
    ax.tick_params(colors=MUTED, labelsize=8)
    
    for spine in ax.spines.values():
        spine.set_edgecolor(GRID)
        spine.set_linewidth(0.8)
    
    if title:  
        ax.set_title(title, color=TEXT, fontsize=9, fontweight='bold', pad=6)
    if xlabel: 
        ax.set_xlabel(xlabel, color=MUTED, fontsize=8)
    if ylabel: 
        ax.set_ylabel(ylabel, color=MUTED, fontsize=8)
    
    ax.grid(color=GRID, linewidth=0.4, alpha=0.6)
```

```python
def page_header(fig, title, subtitle):
    """
    Add a centered header banner to the top of a figure.
    
    Parameters
    ----------
    fig : matplotlib.figure.Figure
        The figure to add header to
    title : str
        Main title text (large, bold)
    subtitle : str
        Subtitle text (smaller, muted)
    """
    fig.text(0.5, 0.985, title, ha='center', va='top',
             fontsize=16, fontweight='bold', color=TEXT, 
             fontfamily='monospace')
    fig.text(0.5, 0.963, subtitle, ha='center', va='top',
             fontsize=8.5, color=MUTED)
```

```python
def save_and_display(fig, filename, chart_num):
    """
    Save figure to PNG AND display it inline in Jupyter.
    
    Parameters
    ----------
    fig : matplotlib.figure.Figure
        The figure to save and display
    filename : str
        Output filename (e.g., 'chart_01_dataset_overview.png')
    chart_num : int
        Chart number for progress tracking
    """
    # Save to file
    fig.savefig(filename, dpi=140, bbox_inches='tight', facecolor=BG)
    print(f"  ✓ Chart {chart_num}: {filename}")
    
    # Display inline in Jupyter
    plt.show()
    
    # Clean up to free memory
    plt.close(fig)
```

```python
# =========================================================================
# CHART 1 — DATASET OVERVIEW
# Shows: Class split, dimensions scatter, megapixel distribution
# =========================================================================

fig1 = plt.figure(figsize=(20, 10))
fig1.patch.set_facecolor(BG)
page_header(fig1,
    "SpectraAgent — Dataset Overview",
    f"Total: {n_total} images  ·  Authentic: {n_auth}  ·  Forged: {n_forg}")

gs1 = GridSpec(2, 3, figure=fig1, hspace=0.35, wspace=0.30,
               top=0.90, bottom=0.08, left=0.06, right=0.96)

# 1A: Class pie chart
ax1a = fig1.add_subplot(gs1[0, 0])
ax1a.set_facecolor(PANEL)
sizes = [n_auth, n_forg]
colors = [AUTH_C, FORG_C]
wedges, texts, autotexts = ax1a.pie(
    sizes, colors=colors, autopct='%1.1f%%',
    startangle=90, pctdistance=0.7,
    wedgeprops=dict(linewidth=3, edgecolor=BG, width=0.5))

for at in autotexts:
    at.set_color(TEXT); at.set_fontsize(11); at.set_fontweight('bold')

ax1a.text(0, 0, f"{n_total}\nimages", ha='center', va='center',
          color=TEXT, fontsize=11, fontweight='bold')
style_ax(ax1a, "Class Distribution")

# 1B: Width vs Height scatter
ax1b = fig1.add_subplot(gs1[0, 1:])
style_ax(ax1b, "Image Dimensions (Width vs Height)", "Width (pixels)", "Height (pixels)")
ax1b.scatter(auth_df['width'], auth_df['height'], c=AUTH_C, s=25, alpha=0.6, 
            label='Authentic', edgecolors='none')
ax1b.scatter(forg_df['width'], forg_df['height'], c=FORG_C, s=25, alpha=0.6, 
            label='Forged', edgecolors='none')
lim_max = max(df['width'].max(), df['height'].max())
ax1b.plot([0, lim_max], [0, lim_max], color=ACC1, lw=1, ls='--', 
         alpha=0.6, label='Square (W=H)')
ax1b.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

# 1C: Megapixels distribution
ax1c = fig1.add_subplot(gs1[1, 0])
style_ax(ax1c, "Megapixel Distribution", "Megapixels", "Count")
bins_mp = np.linspace(0, df['megapixels'].quantile(0.99), 30)
ax1c.hist(auth_df['megapixels'], bins=bins_mp, color=AUTH_C, alpha=0.65, 
         label='Authentic', edgecolor=BG)
ax1c.hist(forg_df['megapixels'], bins=bins_mp, color=FORG_C, alpha=0.65, 
         label='Forged', edgecolor=BG)
ax1c.axvline(auth_df['megapixels'].mean(), color=AUTH_C, lw=1.5, ls='--')
ax1c.axvline(forg_df['megapixels'].mean(), color=FORG_C, lw=1.5, ls='--')
ax1c.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

# 1D: Aspect ratio
ax1d = fig1.add_subplot(gs1[1, 1])
style_ax(ax1d, "Aspect Ratio Distribution", "Aspect Ratio (W/H)", "Count")
bins_ar = np.linspace(0, df['aspect_ratio'].quantile(0.99), 30)
ax1d.hist(auth_df['aspect_ratio'], bins=bins_ar, color=AUTH_C, alpha=0.65, 
         label='Authentic', edgecolor=BG)
ax1d.hist(forg_df['aspect_ratio'], bins=bins_ar, color=FORG_C, alpha=0.65, 
         label='Forged', edgecolor=BG)
ax1d.axvline(1.0, color=ACC1, lw=1.2, ls='--', alpha=0.7, label='Square')
ax1d.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

# 1E: Summary stats
ax1e = fig1.add_subplot(gs1[1, 2])
ax1e.set_facecolor(PANEL); ax1e.axis('off')
summary_text = f"""
DIMENSIONS
  Width:      {df['width'].min():.0f} – {df['width'].max():.0f} px
  Height:     {df['height'].min():.0f} – {df['height'].max():.0f} px
  Megapixels: {df['megapixels'].min():.2f} – {df['megapixels'].max():.2f} MP

AUTHENTIC
  Avg Size:   {auth_df['width'].mean():.0f} × {auth_df['height'].mean():.0f} px
  Avg MP:     {auth_df['megapixels'].mean():.2f} MP

FORGED
  Avg Size:   {forg_df['width'].mean():.0f} × {forg_df['height'].mean():.0f} px
  Avg MP:     {forg_df['megapixels'].mean():.2f} MP
"""
ax1e.text(0.05, 0.95, summary_text, transform=ax1e.transAxes,
         color=TEXT, fontsize=8, fontfamily='monospace', va='top',
         bbox=dict(boxstyle='round,pad=0.8', facecolor='#0d1b2e',
                   edgecolor=GRID, alpha=0.9))

save_and_display(fig1, 'chart_01_dataset_overview.png', 1)
```

```python
# ==========================================================================
# CHART 2 — BRIGHTNESS, CONTRAST & ENTROPY
# Key visual properties that differ between authentic and forged
# ==========================================================================

fig2 = plt.figure(figsize=(20, 12))
fig2.patch.set_facecolor(BG)
page_header(fig2,
    "SpectraAgent — Brightness · Contrast · Entropy",
    "Visual property distributions showing authentic vs forged differences")

gs2 = GridSpec(2, 3, figure=fig2, hspace=0.38, wspace=0.28,
               top=0.91, bottom=0.07, left=0.06, right=0.96)

metrics = [
    ('brightness', 'Brightness Distribution', 'Mean Pixel Value (0-255)'),
    ('contrast', 'Contrast Distribution', 'Std Dev of Pixels'),
    ('entropy', 'Entropy Distribution', 'Shannon Entropy (bits)'),
]

# Top row: Histograms
for i, (metric, title, xlabel) in enumerate(metrics):
    ax = fig2.add_subplot(gs2[0, i])
    style_ax(ax, title, xlabel, "Count")
    
    bins = np.linspace(df[metric].min(), df[metric].max(), 30)
    ax.hist(auth_df[metric], bins=bins, color=AUTH_C, alpha=0.65, 
           label='Authentic', edgecolor=BG)
    ax.hist(forg_df[metric], bins=bins, color=FORG_C, alpha=0.65, 
           label='Forged', edgecolor=BG)
    
    ax.axvline(auth_df[metric].mean(), color=AUTH_C, lw=1.5, ls='--',
              label=f"Auth μ={auth_df[metric].mean():.1f}")
    ax.axvline(forg_df[metric].mean(), color=FORG_C, lw=1.5, ls='--',
              label=f"Forg μ={forg_df[metric].mean():.1f}")
    
    ax.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=7)

# Bottom row: Box plots
for i, (metric, title, _) in enumerate(metrics):
    ax = fig2.add_subplot(gs2[1, i])
    style_ax(ax, f"{title.split()[0]} Box Plot", "Class", metric)
    
    data = [auth_df[metric].dropna(), forg_df[metric].dropna()]
    bp = ax.boxplot(data, patch_artist=True, widths=0.5,
                   medianprops=dict(color=TEXT, linewidth=2))
    
    bp['boxes'][0].set_facecolor(AUTH_C); bp['boxes'][0].set_alpha(0.6)
    bp['boxes'][1].set_facecolor(FORG_C); bp['boxes'][1].set_alpha(0.6)
    
    for element in ['whiskers', 'caps']:
        for item in bp[element]:
            item.set_color(MUTED)
    
    # Overlay points
    for j, (vals, color) in enumerate([(data[0], AUTH_C), (data[1], FORG_C)], 1):
        if len(vals) > 500:
            vals = vals.sample(500)
        jitter = np.random.normal(0, 0.04, len(vals))
        ax.scatter(np.full(len(vals), j) + jitter, vals,
                  c=color, s=8, alpha=0.3, edgecolors='none', zorder=3)
    
    ax.set_xticklabels(['Authentic', 'Forged'], color=MUTED, fontsize=9)

save_and_display(fig2, 'chart_02_brightness_contrast_entropy.png', 2)
```

```python
# ==========================================================================
# CHART 3 — FORENSIC SIGNALS (ELA, Noise, FFT)
# The KEY features for detecting forgeries
# ==========================================================================

fig3 = plt.figure(figsize=(20, 12))
fig3.patch.set_facecolor(BG)
page_header(fig3,
    "SpectraAgent — Forensic Signals Analysis ⭐",
    "ELA · Noise Residual · FFT  —  The strongest forgery indicators")

gs3 = GridSpec(2, 3, figure=fig3, hspace=0.38, wspace=0.28,
               top=0.91, bottom=0.07, left=0.06, right=0.96)

forensic_metrics = [
    ('ela_mean', 'ELA Mean Distribution', 'ELA Mean (JPEG artifacts)'),
    ('noise_mean', 'Noise Residual Mean', 'Noise Mean (high-pass filter)'),
    ('fft_center', 'FFT Center Energy', 'FFT Center Energy (log scale)'),
]

# Top row: Histograms
for i, (metric, title, xlabel) in enumerate(forensic_metrics):
    ax = fig3.add_subplot(gs3[0, i])
    style_ax(ax, title, xlabel, "Count")
    
    bins = np.linspace(df[metric].min(), df[metric].max(), 30)
    ax.hist(auth_df[metric], bins=bins, color=AUTH_C, alpha=0.65, 
           label='Authentic', edgecolor=BG)
    ax.hist(forg_df[metric], bins=bins, color=FORG_C, alpha=0.65, 
           label='Forged', edgecolor=BG)
    
    a_mean = auth_df[metric].mean()
    f_mean = forg_df[metric].mean()
    ratio = f_mean / (a_mean + 1e-10)
    
    ax.axvline(a_mean, color=AUTH_C, lw=1.5, ls='--',
              label=f"Auth μ={a_mean:.3f}")
    ax.axvline(f_mean, color=FORG_C, lw=1.5, ls='--',
              label=f"Forg μ={f_mean:.3f} ({ratio:.2f}×)")
    
    ax.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=7)

# Bottom row: Box plots
for i, (metric, title, _) in enumerate(forensic_metrics):
    ax = fig3.add_subplot(gs3[1, i])
    style_ax(ax, f"{metric.upper()} Statistical Comparison", "Class", metric)
    
    data = [auth_df[metric].dropna(), forg_df[metric].dropna()]
    bp = ax.boxplot(data, patch_artist=True, widths=0.5,
                   medianprops=dict(color=TEXT, linewidth=2), showfliers=False)
    
    bp['boxes'][0].set_facecolor(AUTH_C); bp['boxes'][0].set_alpha(0.6)
    bp['boxes'][1].set_facecolor(FORG_C); bp['boxes'][1].set_alpha(0.6)
    
    for element in ['whiskers', 'caps']:
        for item in bp[element]:
            item.set_color(MUTED)
    
    means = [data[0].mean(), data[1].mean()]
    ax.scatter([1, 2], means, color=ACC1, s=100, zorder=5, 
              marker='D', edgecolors=TEXT, linewidth=1.5, label='Mean')
    
    ax.set_xticklabels(['Authentic', 'Forged'], color=MUTED, fontsize=9)
    ax.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

save_and_display(fig3, 'chart_03_forensic_signals.png', 3)
```

```python
# =========================================================================
# CHART 4 — RGB CHANNEL ANALYSIS
# Per-channel mean and standard deviation
# =========================================================================

fig4 = plt.figure(figsize=(20, 10))
fig4.patch.set_facecolor(BG)
page_header(fig4,
    "SpectraAgent — RGB Channel Analysis",
    "Red · Green · Blue channel statistics by class")

gs4 = GridSpec(1, 3, figure=fig4, hspace=0.35, wspace=0.28,
               top=0.88, bottom=0.12, left=0.06, right=0.96)

channels = [
    ('r_mean', 'Red Channel Mean', CH_R),
    ('g_mean', 'Green Channel Mean', CH_G),
    ('b_mean', 'Blue Channel Mean', CH_B),
]

for i, (metric, title, color) in enumerate(channels):
    ax = fig4.add_subplot(gs4[0, i])
    style_ax(ax, title, "Class", "Mean Pixel Value (0-255)")
    
    data = [auth_df[metric].dropna(), forg_df[metric].dropna()]
    bp = ax.boxplot(data, patch_artist=True, widths=0.5,
                   medianprops=dict(color=TEXT, linewidth=2))
    
    for box in bp['boxes']:
        box.set_facecolor(color); box.set_alpha(0.6)
    
    for element in ['whiskers', 'caps']:
        for item in bp[element]:
            item.set_color(MUTED)
    
    for j, (vals, jitter_c) in enumerate([(data[0], AUTH_C), (data[1], FORG_C)], 1):
        if len(vals) > 500:
            vals = vals.sample(500)
        jitter = np.random.normal(0, 0.04, len(vals))
        ax.scatter(np.full(len(vals), j) + jitter, vals,
                  c=jitter_c, s=12, alpha=0.4, edgecolors='none', zorder=3)
    
    ax.set_xticklabels(['Authentic', 'Forged'], color=MUTED, fontsize=9)
    
    a_m = data[0].mean()
    f_m = data[1].mean()
    ax.text(0.5, 0.95, f"Δ = {abs(f_m - a_m):.1f}", 
           transform=ax.transAxes, ha='center', va='top',
           color=ACC1, fontsize=10, fontweight='bold',
           bbox=dict(boxstyle='round,pad=0.3', facecolor=PANEL, 
                    edgecolor=ACC1, alpha=0.8))

save_and_display(fig4, 'chart_04_rgb_channels.png', 4)
```

```python
# =========================================================================
# CHART 5 — STATISTICAL COMPARISON (Multiple Metrics)
# Side-by-side box plots for quick comparison
# =========================================================================

fig5 = plt.figure(figsize=(20, 10))
fig5.patch.set_facecolor(BG)
page_header(fig5,
    "SpectraAgent — Statistical Comparison",
    "Key metrics: Authentic vs Forged side-by-side")

gs5 = GridSpec(2, 4, figure=fig5, hspace=0.38, wspace=0.30,
               top=0.90, bottom=0.08, left=0.06, right=0.96)

comparison_metrics = [
    'brightness', 'contrast', 'entropy', 'noise_mean',
    'ela_mean', 'r_mean', 'g_mean', 'b_mean'
]

for i, metric in enumerate(comparison_metrics):
    row = i // 4
    col = i % 4
    ax = fig5.add_subplot(gs5[row, col])
    style_ax(ax, metric.replace('_', ' ').title(), "Class", metric)
    
    data = [auth_df[metric].dropna(), forg_df[metric].dropna()]
    bp = ax.boxplot(data, patch_artist=True, widths=0.6,
                   medianprops=dict(color=TEXT, linewidth=2))
    
    bp['boxes'][0].set_facecolor(AUTH_C); bp['boxes'][0].set_alpha(0.6)
    bp['boxes'][1].set_facecolor(FORG_C); bp['boxes'][1].set_alpha(0.6)
    
    for element in ['whiskers', 'caps']:
        for item in bp[element]:
            item.set_color(MUTED)
    
    ax.set_xticklabels(['Auth', 'Forg'], color=MUTED, fontsize=8)

save_and_display(fig5, 'chart_05_statistical_comparison.png', 5)
```

```python
# =========================================================================
# CHART 6 — CORRELATION HEATMAP
# Shows which features correlate with each other
# =========================================================================

fig6 = plt.figure(figsize=(16, 14))
fig6.patch.set_facecolor(BG)
page_header(fig6,
    "SpectraAgent — Feature Correlation Matrix",
    "Pearson correlation between numerical features")

gs6 = GridSpec(1, 1, figure=fig6, top=0.92, bottom=0.05, left=0.05, right=0.95)
ax6 = fig6.add_subplot(gs6[0, 0])

corr_features = [
    'width', 'height', 'megapixels', 'aspect_ratio',
    'brightness', 'contrast', 'entropy',
    'r_mean', 'g_mean', 'b_mean',
    'noise_mean', 'ela_mean', 'fft_center'
]

corr = df[corr_features].corr()

if HAS_SEABORN:
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn_r',
               center=0, square=True, linewidths=0.5,
               cbar_kws={'shrink': 0.8}, ax=ax6,
               annot_kws={'size': 7})
    ax6.set_facecolor(PANEL)
    ax6.tick_params(colors=TEXT, labelsize=8)
else:
    im = ax6.imshow(corr, cmap='RdYlGn_r', vmin=-1, vmax=1, aspect='auto')
    ax6.set_xticks(np.arange(len(corr_features)))
    ax6.set_yticks(np.arange(len(corr_features)))
    ax6.set_xticklabels(corr_features, rotation=45, ha='right', fontsize=8, color=TEXT)
    ax6.set_yticklabels(corr_features, fontsize=8, color=TEXT)
    
    for i in range(len(corr_features)):
        for j in range(len(corr_features)):
            text = ax6.text(j, i, f'{corr.iloc[i, j]:.2f}',
                          ha="center", va="center", color=TEXT, fontsize=6)
    
    plt.colorbar(im, ax=ax6, shrink=0.8)
    ax6.set_facecolor(PANEL)

save_and_display(fig6, 'chart_06_correlation_heatmap.png', 6)
```

```python
# =========================================================================
# CHART 7 — DIMENSION ANALYSIS
# Detailed look at image sizes and aspect ratios
# =========================================================================

fig7 = plt.figure(figsize=(20, 10))
fig7.patch.set_facecolor(BG)
page_header(fig7,
    "SpectraAgent — Image Dimension Analysis",
    "Resolution, aspect ratio, and size distributions")

gs7 = GridSpec(1, 3, figure=fig7, hspace=0.35, wspace=0.28,
               top=0.88, bottom=0.12, left=0.06, right=0.96)

# 7A: Megapixels
ax7a = fig7.add_subplot(gs7[0, 0])
style_ax(ax7a, "Megapixels Distribution", "Megapixels", "Count")
bins_mp = np.linspace(0, df['megapixels'].quantile(0.99), 25)
ax7a.hist(auth_df['megapixels'], bins=bins_mp, color=AUTH_C, alpha=0.65, 
         label='Authentic', edgecolor=BG)
ax7a.hist(forg_df['megapixels'], bins=bins_mp, color=FORG_C, alpha=0.65, 
         label='Forged', edgecolor=BG)
ax7a.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

# 7B: Aspect ratio
ax7b = fig7.add_subplot(gs7[0, 1])
style_ax(ax7b, "Aspect Ratio Distribution", "Aspect Ratio (W/H)", "Count")
bins_ar = np.linspace(0, df['aspect_ratio'].quantile(0.99), 25)
ax7b.hist(auth_df['aspect_ratio'], bins=bins_ar, color=AUTH_C, alpha=0.65, 
         label='Authentic', edgecolor=BG)
ax7b.hist(forg_df['aspect_ratio'], bins=bins_ar, color=FORG_C, alpha=0.65, 
         label='Forged', edgecolor=BG)
ax7b.axvline(1.0, color=ACC1, lw=1.2, ls='--', label='Square (AR=1)')
ax7b.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

# 7C: Scatter
ax7c = fig7.add_subplot(gs7[0, 2])
style_ax(ax7c, "Megapixels vs Aspect Ratio", "Megapixels", "Aspect Ratio")
ax7c.scatter(auth_df['megapixels'], auth_df['aspect_ratio'],
            c=AUTH_C, s=25, alpha=0.5, label='Authentic', edgecolors='none')
ax7c.scatter(forg_df['megapixels'], forg_df['aspect_ratio'],
            c=FORG_C, s=25, alpha=0.5, label='Forged', edgecolors='none')
ax7c.legend(facecolor=PANEL, edgecolor=GRID, labelcolor=TEXT, fontsize=8)

save_and_display(fig7, 'chart_07_dimension_analysis.png', 7)
```

```python
# =========================================================================
# COMPLETION SUMMARY
# =========================================================================

print("\n" + "="*70)
print("  SECTION 3 COMPLETE")
print("="*70)
print(f"\n  Generated and displayed 7 charts:")
print(f"    ✓ All charts displayed inline above")
print(f"    ✓ All charts saved as PNG files")
print(f"\n  Key Findings:")

ela_ratio = forg_df['ela_mean'].mean() / (auth_df['ela_mean'].mean() + 1e-10)
ent_ratio = forg_df['entropy'].mean() / (auth_df['entropy'].mean() + 1e-10)
noise_ratio = forg_df['noise_mean'].mean() / (auth_df['noise_mean'].mean() + 1e-10)

print(f"    • ELA Mean:     {ela_ratio:.2f}× higher in forged")
print(f"    • Entropy:      {ent_ratio:.2f}× higher in forged")
print(f"    • Noise Mean:   {noise_ratio:.2f}× higher in forged")
print(f"\n  These 3 signals are the strongest forgery indicators!")
print("="*70 + "\n")
```
